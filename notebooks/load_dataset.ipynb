```python
from datasets import load_dataset
import os
import openai
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
dataset = load_dataset("imagefolder", data_dir="../scripts/talks_dset/")
```

Let's have a look at a galaxy

```python
dataset['train'][2]['image']
```

```python
# Original conversation
print(dataset['train'][2]['conversation'])
```

```python
# Generate a summary of the conversation in the form of an answer
summary = openai.Completion.create(
  model="text-davinci-003",
  prompt="""Given the following conversation between galaxy zoo users that comment on an astronomical image, answer the following question about this image making sure to include as much information as possible.

Conversation:
---
%s
---

Question: What do we see in this image?
Answer:"""%dataset['train'][2]['conversation'],
  max_tokens=256
).choices[0].text
print(summary)
```
